```python
data2=pd.DataFrame((data['CDRGLOB'],data['APA'],data['INDEPEND'],data['CSFVOL'], data["NACCMMSE"])).T
data2
```

```python
data2['NORMD_CSFVOL']=(data2['CSFVOL']-data2['CSFVOL'].min())/(data2['CSFVOL'].max()-data2['CSFVOL'].min())
data2
```

Obtaining and normalizing the formula

```python
data["Formula"] = ((5-(data2["NORMD_CSFVOL"]))*(5-(data2["APA"]))*(5-(data2["INDEPEND"])))*(8-data2["NACCMMSE"])

data2['NORMD_Formula']=(data['Formula']-data['Formula'].min())/(data['Formula'].max()-data['Formula'].min())

correlation = data["Formula"].corr(data["CDRGLOB"])
correlation
```
